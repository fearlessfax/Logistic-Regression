```python
# 1. initalize Weight

def initialize_weight():
    W = np.zeros(784) + 0.001
    B = 0.1
    return W,B

# 2. Weighted Sum Calculation

def weighted_sum(W,B,X):
    weighted_sum = W.dot(X) + B
    return weighted_sum
    
# 3. Sigmoid Calculation

def sigmoid(X):
    return 1 / (1 + np.exp(-X))

def Sigmoid_calculation(Weighted_sum):
    cal_sig = sigmoid(Weighted_sum)
    return cal_sig

def predict(Y):
    return np.where(Y>0.5,1,0)

def forward(W,B,X):
    weight_sum = weighted_sum(W,B,X)
    Y = Sigmoid_calculation(weight_sum)
    return Y

def gradient(Y,Y_label,X):
    M = len(Y)
    dW = X.dot(Y - Y_label)/M
    dB = (Y - Y_label)/m
    dB = dB.sum()
    return dW , dB
    
def update_parameters(learning_rate,W,B,dW,dB):
    W = W - learning_rate*dW
    B = B - learning_rate*dB
    return W,B

def accuracy(Y,Y_train):
    Y_predict = predict(Y)
    Accuracy = np.where(Y_predict == Y_train, 1, 0)
    count_of_correct = np.count_nonzero(Accuracy == 1)
    return count_of_correct/len(Y)
    
```

```python
W,B = initialize_weight()
ww = weighted_sum(W,B,X_train)
Y = forward(W,B,X_train)
DW,DB = gradient(Y,Y_train,X_train)
print(Y)
```

```python
def model_fit(X,Y,alpha,iteration):
    
    W,B = initialize_weight()

    for i in range(iteration):
        Y = forward(W,B,X)
        dW, dB = gradient(Y,Y_train,X_train)
        W,B = update_parameters(alpha,W,B,dW,dB)
        print(accuracy(Y,Y_train))
         
    return W,B
```

```python
W,B = model_fit(X_train,Y_train,0.1,100)
```
